```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# 1. Create dummy sales data
data = {'Month': [1, 2, 3, 4, 5, 6],
        'Sales': [100, 120, 130, 150, 170, 200]}
df = pd.DataFrame(data)

# 2. Prepare data for model
X = df[['Month']]
y = df['Sales']

# 3. Build and train a simple model
model = LinearRegression()
model.fit(X, y)

# 4. Predict next month (Month 7)
next_month = np.array([[7]])
prediction = model.predict(next_month)

print(f"Predicted sales for Month 7: {prediction[0]}")

# 5. Visualize the results
plt.scatter(df['Month'], df['Sales'], color='blue', label='Actual')
plt.plot(df['Month'], model.predict(X), color='red', label='Trend')
plt.scatter(7, prediction, color='green', label='Prediction')
plt.legend()
plt.show()
```
